```python
#creating predictors for mL
#convert venue column into categorical so its a number value

matches['venue_code']= matches['venue'].astype('category').cat.codes  

#we created a col called venue code for the categorical value
```

```python
matches
```

```python
matches['opp_code'] = matches['opponent'].astype('category').cat.codes
```

```python
matches
```

```python
#converting the time in time column (prev format is h:m but were changing to just hour)
#the code takes time column only keeps hour and leaves blank for minute

matches['hour'] = matches['time'].str.replace(':.+', '', regex=True).astype('int')  
```

```python
matches
```

```python
matches['day_code']= matches['date'].dt.dayofweek
```

```python
#were creating a target col that only give 1 to wins and 0 for both loss and draw just for simplicity

matches['target'] = (matches['result'] == 'W').astype('int')
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# random forest is a series of decision tress
# n_estimators is number or branches individidual decision tress with slighlty different parameters
#the higher n_estimators is longer itll run and more accurate itll be
# min_sample_split is number of samples before we split the node of the tree 
#the higher min_sample_split is the lower the accuracy  

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches[matches['date'] < '2022-01-01']
```

```python
test = matches[matches['date'] > '2022-01-01']
```

```python
predictors = ['venue_code', 'opp_code', 'hour', 'day_code']
```

```python
# rf fit takes sample from predictors and predicts outcome 1 for win or 0 for loss/draw

rf.fit(train[predictors], train['target'])
```

```python
preds = rf.predict(test[predictors])
```

```python
# below calculates what prediction of the time your prediction was accurate

from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test['target'], preds)
```

```python
# we can see 61% of the time our accuracy was correct

acc
```

```python
def rolling_averages(group, cols, new_cols):
    group['date'] = pd.to_datetime(group['date'], errors = 'coerce')
    group = group.sort_values('date')
    group = group.set_index('date')
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.reset_index()
    group = group.dropna(subset=new_cols)   #removes rows with missing values
    return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f'{c}_rolling' for c in cols]
```

```python
new_cols   #new columns with rolling averages
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
# dropping the team level on the left side

matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
#unique values for indexes so we get unqiue values

matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
#Retraining our ML

def make_predictions(data, predictors):
    train = data[data['date'] < '2022-01-01']
    test = data[data['date'] > '2022-01-01']
    rf.fit(train[predictors], train['target'])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test['target'], predicted = preds), index=test.index)
    precision = precision_score(test['target'], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```

```python
#adding the other columns by merging them based on index

combined = combined.merge(matches_rolling[['date', 'team', 'opponent', 'result']], left_index=True, right_index=True)
```

```python
#we need to make sure the team names are the same for both team and opponent columns as theyre different for ex: wolves & wolverhampton wanderes

class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {
    "Brighton and Hove Albion": "Brighton", 
    "Manchester United": "Manchester Utd", 
    "Newcastle United": "Newcastle Utd", 
    "Tottenham Hotspur": "Tottenham", 
    "West Ham United": "West Ham", 
    "Wolverhampton Wanderers": "Wolves"

} 

mapping = MissingDict(**map_values)
```

```python
mapping['West Ham United']
```

```python
combined['new_team'] = combined['team'].map(mapping)
```

```python
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
```
